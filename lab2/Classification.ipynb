```python
not_need = ['chol', 'fbs']

x = df.drop(not_need, axis=1)
x.head()
```

```python
plt.scatter(x=x.age[x.target==1], y=x.thalach[(x.target==1)], c="red")
plt.scatter(x=x.age[x.target==0], y=x.thalach[(x.target==0)])
plt.legend(["Disease", "Not Disease"])
plt.xlabel("Age")
plt.ylabel("Maximum Heart Rate")
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10,6))
ax = sns.distplot(x['age'], bins=10)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.boxplot(x=x["age"])
plt.show()
```

```python
young_ages = x[(x.age>=29) & (x.age<40)] 
middle_ages = x[(x.age>=40) & (x.age<55)]
elderly_ages = x[(x.age>55)]
print('Young Ages :',len(young_ages))
print('Middle Ages :',len(middle_ages))
print('Elderly Ages :',len(elderly_ages))
```

```python
  #we can see in pie.

plt.figure(figsize = (10,10))
plt.pie([len(young_ages),len(middle_ages),len(elderly_ages)],labels=['young ages','middle ages','elderly ages'])
plt.title('Age States')
plt.show()
```
