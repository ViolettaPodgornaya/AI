```python
names = ['alfa', 'audi', 'bmw', 'chevrolet', 'dodge', 'honda', 'isuzu', 'jaguar', 'maxda', 'mazda', 'buick', 
        'mercury', 'mitsubishi', 'Nissan', 'nissan', 'peugeot', 'plymouth', 'porsche', 'porcshce', 'renault', 'saab',
        'subaru', 'toyota', 'toyouta', 'vokswagen', 'volkswagen', 'vw', 'volvo']

un_names = df.CarName.unique()
x = df.copy()

for i, name in enumerate(names):
    for un_name in un_names:
        if name in un_name:
            x['CarName'][x['CarName'] == un_name] = i

x.CarName.unique()
```

```python
x = x.drop(['car_ID'],axis=1)
x.head()
```

```python
x.fueltype.unique()
```

```python
x['fueltype'].replace(x.fueltype.unique(), [1, 0], inplace=True)
x.head()
```

```python
x.enginelocation.unique()
```

```python
x['enginelocation'].replace(x.enginelocation.unique(), [1, 0], inplace=True)
x.head()
```

```python
x.carbody.unique()
```

```python
x['carbody'].replace(x.carbody.unique(), [i for i in range(len(x.carbody.unique()))], inplace=True)
x.head()
```

```python
x.info()
```

```python
x['CarName'] = x['CarName'].astype('int32')

not_need = ['aspiration', 'doornumber', 'drivewheel', 'enginetype', 'cylindernumber', 'fuelsystem']
```

```python
x = x.drop(not_need, axis=1)
x.head()
```

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(x.corr(),cmap='gist_heat_r')
```
